```python
#Compare and contrast mutable and immutable objects in Python with examples.

Ans -
Objects whose state or values can be changed after they are created is called as mutable objects, example List 

#code list_cont[0]= "pwskills"

and lets say b = "pwskills"
b
insite- If you want to change the b[0] that is b to I, you cannot change it that is called as immutable objects.

Comparison Summary

Modification - 
Mutable: Can be changed in place.
Immutable: Cannot be changed; changes result in a new object.


Performance -
Mutable: Can be more efficient for repeated modifications (e.g., lists).
Immutable: Can be less efficient for modifications but can be used for ensuring data integrity (e.g., strings).

Usage -
Mutable: Suitable for data structures that need to be updated frequently.
Immutable: Suitable for fixed data structures or when you want to ensure that data cannot be altered.
```
